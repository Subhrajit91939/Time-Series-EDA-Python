```python
# Importing the necessary Libraries
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# get TESLA data from yahoofinance
# 5 years data for TESLA from the yahoofinance website.
df_tesla = pdr.get_data_yahoo('TSLA')
```

### **Time Resampling**

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
# This will give the min value of all the features for any particular year in the given dataset.
# For more info and rules: https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
```

```python
df_tesla.resample(rule='A').max()
```

```python
# Year end frequency
df_tesla.resample(rule='A').max()['Open'].plot()
sns.set_theme(style='white')
sns.set_style("darkgrid")
```

```python
# quarterly start frequency
df_tesla.resample(rule='QS').max()['Open'].plot()
sns.set_theme(style='white')
sns.set_style("darkgrid")
```

```python
# Business Year end frequency
df_tesla.resample(rule='BA').max()['Open'].plot()
sns.set_theme(style='white')
sns.set_style("darkgrid")
```

```python
df_tesla.resample(rule='A').max()['Open'].plot(kind='bar', figsize=(10,6))
```

```python
df_tesla['Open'].resample(rule='M').mean().plot(kind='bar', figsize=(18,6))
```

```python
# Rolling and Expanding:
df_tesla['High'].rolling(10).mean() 
# Calculating the mean of 10 consecutive data in the 'High' column.
```

```python
df_tesla['Open:30 Days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla[['Open', 'Open:30 Days rolling']].plot(figsize=(15,8))
```

```python
df_microsoft = pdr.get_data_yahoo('MSFT')
```

```python
df_microsoft.head(10)
```

```python
df_microsoft[:].plot(subplots=True, figsize= (15,20))
```

```python
# Day the microsoft stock hit its maximum values
max_open = df_microsoft.max()['Open']
print(f"Maximum Stock Open Price for Microsoft: {max_open}")
print(f"Date of the Max Stock Open price: {df_microsoft[(df_microsoft['Open'] == max_open)].reset_index()['Date'][0]}")
```

```python
# Day the microsoft stock hit its maximum values
min_open = df_microsoft.min()['Open']
print(f"Minimum Stock Open Price for Microsoft: {min_open}")
print(f"Date of the Min Stock Open price: {df_microsoft[(df_microsoft['Open'] == min_open)].reset_index()['Date'][0]}")
```
